#### Mission - Sandviç Kombinasyonları
Tüm ekmek, et, sebze ve sos kombinasyonlarını üretmek için iç içe `for` döngüleri kullanılır.


```python
# Sandviç kombinasyonları
breads = ["Rye bread", "Wheat", "White"]
meats = ["Meatball", "Sausage", "Chicken breast"]
vegis = ["Lettuce", "Tomato", "Cucumber"]
sauces = ["Mayonnaise", "Honey Mustard", "Chili"]

print("Possible combinations of David's sandwich shop")
for b in breads:
    for m in meats:
        for v in vegis:
            for s in sauces:
                print(b + " + " + m + " + " + v + " + " + s)
```

#### 1. for Döngüsünün Temelleri
`for` döngüsü tekrarlayan işlemler için kullanılır. Döngü kontrol değişkeni (`i`) her iterasyonda farklı değer alır. Kullanılmayan değişkenler için `_` (underscore) kullanılabilir.


```python
# Tekrarlayan print örneği (döngü olmadan)
print("Welcome to everyone!!")
print("Welcome to everyone!!")
print("Welcome to everyone!!")
print("Welcome to everyone!!")
print("Welcome to everyone!!")

# for döngüsü ile
for i in range(5):
    print("Welcome to everyone!!")

# Döngü kontrol değişkenini kullanma
for i in range(5):
    print(i, "Welcome to everyone!!")

# Kullanılmayan değişken için _ kullanımı
for _ in range(5):
    print("Welcome to everyone!!")
```

#### 2. range() Fonksiyonu
`range(start, end, step)` belirli bir aralıktaki sayıları üretir. `start` varsayılan 0, `step` varsayılan 1'dir. `end` değeri dahil değildir. Sadece `end` verilirse 0'dan başlar. `range()` sadece tam sayılarla çalışır.


```python
# range() örnekleri
print(list(range(0, 5, 1)))  # [0, 1, 2, 3, 4]
print(list(range(0, 5, 2)))  # [0, 2, 4]
print(list(range(2, 5)))     # [2, 3, 4] - step varsayılan 1
print(list(range(0, 10, 2))) # [0, 2, 4, 6, 8]
print(list(range(5)))        # [0, 1, 2, 3, 4] - start varsayılan 0

# range() ile toplam hesaplama
s = 0
for i in range(1, 11):  # 1'den 10'a kadar
    s += i
print("Sum of numbers from 1 to 10:", s)
print(sum(range(1, 11)))  # Aynı sonuç

# Faktöriyel hesaplama
n = int(input("Enter a number: "))
fact = 1
for i in range(1, n + 1):
    fact = fact * i
print(str(n) + "! =", fact)

# Hata örneği - float kullanımı
try:
    for i in range(0.9):
        print(i)
except TypeError as e:
    print("Hata:", e)
```

#### 3. Dizi Nesneleri Üzerinde Iterasyon
Liste, string gibi sıralı (sequence) nesneler `for` döngüsü ile doğrudan iterasyon yapılabilir. String karakter karakter, liste eleman eleman dolaşılır.

#### 4. Pratik Uygulamalar ve format() Fonksiyonu
`format()` fonksiyonu çıktıyı biçimlendirmek için kullanılır. `{:5d}` tam sayı için 5 karakter genişlik, `{:6.3f}` ondalıklı sayı için 6 karakter genişlik ve 3 ondalık basamak belirtir. Snake matrix gibi karmaşık desenler iç içe döngülerle oluşturulur.


```python
# format() örnekleri
print("{:5d}".format(100))   # Tam sayı formatı - 5 karakter genişlik
print("{:5d}".format(10))    # Tam sayı formatı
print("{:6.3f}".format(3.14)) # Ondalıklı sayı formatı - 6 karakter, 3 ondalık
print("{:6.2f}".format(3.14)) # Ondalıklı sayı formatı - 6 karakter, 2 ondalık

# Çift sayıların toplamı (1-100)
print("Sum of even from 1 to 100 :", sum(range(0, 101, 2)))

# Tek sayıların toplamı (1-100)
print("Sum of odd from 1 to 100 :", sum(range(1, 100, 2)))

# range() ile geriye doğru sayma
for i in range(10, 0, -1):  # 10'dan 1'e kadar
    print(i, end=" ")
print()

# Liste ile for döngüsü
for i in [1, 2, 3, 4, 5]:  # Liste doğrudan kullanılabilir
    print("Welcome.")

# String ile for döngüsü
for i in "Hello":
    print("i =", i)

# range() ile çoklu satır
for i in range(5):
    print("Welcome to")
    print("Python corporation!")

# range() örnekleri
for i in range(1, 6, 1):  # 1'den 5'e kadar
    print(i, end=" ")
print()

for i in range(1, 10, 2):  # 1'den 9'a kadar, 2'şer artış
    print(i, end=" ")
print()

for i in range(10, 0, -1):  # 10'dan 1'e kadar, geriye
    print(i, end=" ")
print()

# Snake Matrix örneği (n x n koltuk düzeni)
# Yılan şeklinde artan sayılar: çift satırlar soldan sağa, tek satırlar sağdan sola
n = int(input("Enter n :"))
seat_list, element = [], 0

for i in range(n):
    temp, count = [], 0
    for j in range(n):
        element += 1
        temp.append(element)
        count += 1
        if count == n:
            seat_list.append(temp)

for i in range(len(seat_list)):
    if i % 2 == 0:  # Çift satırlar (0, 2, 4...) soldan sağa
        for j in range(len(seat_list[i])):
            print("{:2d}".format(seat_list[i][j]), end=" ")
        print()
    else:  # Tek satırlar (1, 3, 5...) sağdan sola
        for j in range(len(seat_list[i]) - 1, -1, -1):
            print("{:2d}".format(seat_list[i][j]), end=" ")
        print()
```

#### Mission - Salyangoz Duvar Problemi
Salyangoz günde 7m yukarı çıkar, 5m aşağı kayar. 30m yüksekliğindeki duvarı kaç günde çıkar? `while` döngüsü ile koşul sağlanana kadar tekrar edilir.


```python
# Salyangoz duvar problemi
current_pos = 0
days = 1
move_up = 7
move_down = 5
destination = 30
current_pos += move_up

print("day :", days, "Snail's position :", current_pos, "meters")
while current_pos <= destination:
    current_pos -= move_down
    days += 1
    current_pos += move_up
    print("day :", days, "Snail's position :", current_pos, "meters")
print()
print("It took {} days to get out of the wall.".format(days))
```

#### 1. while Döngüsü
`while` döngüsü koşul True olduğu sürece çalışır. Tekrar sayısı belirsiz olduğunda veya koşul tabanlı döngülerde kullanılır. `for` döngüsü tekrar sayısı belli olduğunda daha uygundur.


```python
# while döngüsü temel örnek
i = 0  # başlangıç değeri
while i < 5:
    print('Welcome to everyone')
    i += 1

# 1'den n'ye kadar toplam
n = int(input("Enter a number to sum up to :"))
s = 0
i = 1
while i <= n:
    s = s + i
    i += 1
print('Sum of 1 to {} is {}'.format(n, s))

# for ile karşılaştırma
s = 0
for i in range(1, 11):
    s += i
print('Sum of 1 to 10 is', s)

# while ile kullanıcı girişi kontrolü
selected = None
while selected not in ['scissors', 'rock', 'paper']:
    selected = input('Choose among scissors, rock, paper> ')
print('Choose value :', selected)

# Negatif sayı kontrolü ile while
n = -1
while n <= 0:  # Pozitif değer girilene kadar tekrar et
    n = int(input("Enter a number to sum up to :"))
s = 0
for i in range(1, n + 1):
    s = s + i
print('Sum of 1 to {} is {}'.format(n, s))

# Şifre kontrolü
password = ""
while password != "pythonisfun":
    password = input("Enter the password")
print("** Login Success **")

# Koşul tabanlı while (yapı kontrolü)
under_construction = True
while under_construction:
    response = input("Is the construction completed?")
    if response == "Yes":
        under_construction = False
print("Escaped successfully.")

# Belirli sayıda tekrar için while
count = 1
s = 0  # Biriken değerleri tutan değişken, 0 olarak sıfırla
while count <= 10:
    s += count  # Her seferinde count değerini s'ye ekle
    count += 1  # Her seferinde count değerini 1 artır
print('Sum is', s)

# Kullanıcıdan sayılar alarak toplam
total = 0
answer = 'yes'
while answer == 'yes':
    num = int(input("Enter the number: "))
    total += num
    answer = input('Continue?(yes/no): ')
print('Sum is :', total)
```

#### 2. random Modülü Detayları
`random` modülü rastgele sayılar ve seçimler üretmek için kullanılır. `random.random()` 0-1 arası rastgele float, `random.randint(a, b)` a-b arası (dahil) tam sayı, `random.randrange(start, end, step)` belirli aralıkta sayı üretir.


```python
# random modülü fonksiyonları
import random as rd

print(rd.random())              # 0-1 arası rastgele float
print(rd.random())
print(rd.randrange(1, 7))       # 1-6 arası tam sayı
print(rd.randrange(0, 10, 2))   # 0-9 arası çift sayılar
print(rd.randint(1, 10))        # 1-10 arası (dahil) tam sayı

# shuffle() - listeyi karıştırır
numlist = [10, 20, 30, 40, 50]
rd.shuffle(numlist)
print(numlist)

# choice() - listeden rastgele bir eleman seçer
print(rd.choice(numlist))

# sample() - listeden rastgele n eleman seçer
print(rd.sample(numlist, 3))

# Örnek: 1-10 arası sayıları karıştırma
a = list(range(1, 11))
rd.shuffle(a)
print(a)
```

#### 3. break ve continue İfadeleri
`break` döngüyü tamamen sonlandırır, `continue` ise mevcut iterasyonu atlayıp bir sonraki iterasyona geçer. `break` döngüden çıkar, `continue` döngüyü devam ettirir.


```python
# break örneği - sesli harf bulunca dur
st = 'Programming'
for ch in st:
    if ch in ['a', 'e', 'i', 'o', 'u']:
        break  # Sesli harf bulunca döngüden çık
    print(ch)
print('The end')

# continue örneği - sesli harfleri atla
st = 'Programming'
for ch in st:
    if ch in ['a', 'e', 'i', 'o', 'u']:
        continue  # Sesli harfi atla, devam et
    print(ch)
print('The end')
```

#### 4. İç İçe Döngüler (Double Loop)
Bir döngü başka bir döngünün içinde olabilir. Çarpım tablosu, matris işlemleri gibi problemlerde kullanılır. İç döngü dış döngünün her iterasyonunda tamamen çalışır.


```python
# Çarpım tablosu (for ile)
for i in range(2, 10):
    for j in range(1, 10):
        print('{}x{} = {:2d},'.format(i, j, i * j), end=" ")
    print()

# Çarpım tablosu (while ile - 2'nin katları)
i = 2
while i < 3:
    j = 1
    while j < 10:
        print('{} * {} = {}'.format(i, j, i * j))
        j += 1
    print()
    i += 1

# Çarpım tablosu (while ile - 1'den 9'a kadar, break örneği)
i = 1
while i < 3:
    j = 1
    while j < 10:
        if i == 2 and j == 5:
            break  # İç döngüden çık
        print('{}*{}={}'.format(i, j, i * j))
        j += 1
    i += 1
```

#### 5. Pratik Uygulamalar

```python
# Palindrom sayı kontrolü
num = int(input("Enter an integer: "))
temp = num
reverse_num = 0
while(num > 0):
    digit = num % 10
    reverse_num = reverse_num * 10 + digit
    num = num // 10

if(temp == reverse_num):
    print("{} is a palindrome number".format(temp))
else:
    print("{} is not a palindrome number".format(temp))

# Sayı tahmin oyunu
import random

num = random.randint(1, 100)
guess_count = 0

while True:
    print("Guess a number between 1 to 100")
    guess = int(input("Enter a number: "))
    guess_count += 1

    if guess == num:
        print("Congratulations. Total Try = {}".format(guess_count))
        break
    elif guess < num:
        print("Higher!")
    else:
        print("Lower!")
```

## 1. Liste Tanımlama ve Kullanımı

## 2. Liste Elemanlarını Ekleme ve Silme
Listeye eleman eklemek için `+`, `append()`, `extend()`, `insert()` metodları kullanılır. Silmek için `del`, `pop()`, `remove()` metodları kullanılır.


```python
# + operatörü ile liste birleştirme
person1 = ['David Doe', 20, 1, 180.0, 100.0]
person2 = ['John Smith', 25, 1, 170.0, 70.0]
person_list = person1 + person2  # İki listenin elemanlarını birleştir
print(person_list)

# append() - sona eleman ekleme
a_list = ['a', 'b', 'c', 'd', 'e']
a_list.append('f')  # 'f' ekle
print(a_list)  # ['a', 'b', 'c', 'd', 'e', 'f']

n_list = [10, 20, 30, 40]
n_list.append(50)  # 50 ekle
print(n_list)  # [10, 20, 30, 40, 50]

# extend() - liste elemanlarını ekleme
list1 = ['a', 'b', 'c']
list2 = [1, 2, 3]
list1.extend(list2)
print(list1)  # ['a', 'b', 'c', 1, 2, 3]

# extend() string ile kullanımı
list1.extend('d')  # String karakter karakter eklenir
print(list1)  # ['a', 'b', 'c', 1, 2, 3, 'd']

# append() vs extend() farkı
list1 = [11, 22, 33, 44]
list1.append([55, 66])  # Liste olarak ekler (iç içe liste)
print(list1)  # [11, 22, 33, 44, [55, 66]]

list1 = [11, 22, 33, 44]
list1.extend([55, 66])  # Elemanları ekler
print(list1)  # [11, 22, 33, 44, 55, 66]

# insert() - belirli konuma ekleme
slist = ['David', 178.9, 'John', 173.5, 'Jane', 176.1]
print(slist)
slist.insert(4, 'Peter')  # 4. indekse 'Peter' ekle
slist.insert(5, 168.1)    # 5. indekse 168.1 ekle
print(slist)
```

```python
# del - indeks ile silme
n_list = [11, 22, 33, 44]
print(n_list)  # Tüm elemanları yazdır
del n_list[3]  # 44'ü sil
print(n_list)  # [11, 22, 33]

# pop() - son elemanı sil ve döndür
n_list = [10, 20, 30]
print(n_list)  # Tüm elemanları yazdır
n = n_list.pop()  # Son elemanı sil ve döndür
print('n =', n)  # 30
print('n_list =', n_list)  # [10, 20]

# remove() - değere göre silme
n_list = [11, 22, 33, 44, 55, 66]
print(n_list)
n_list.remove(44)
print(n_list)  # [11, 22, 33, 55, 66]

# remove() hata örneği
try:
    n_list = [11, 22, 33, 44, 55, 66]
    n_list.remove(88)  # 88 listede yok
except ValueError as e:
    print("Hata:", e)

# Güvenli silme - in operatörü ile kontrol
n_list = [11, 22, 33, 44, 55, 66]
if 55 in n_list:
    n_list.remove(55)
if 88 in n_list:
    n_list.remove(88)
print(n_list)  # [11, 22, 33, 44, 66]

# BTS örneği
bts = ["V", "J-Hope", "Suga", "Jungkook"]
bts.remove("Jungkook")
print(bts)  # ["V", "J-Hope", "Suga"]

if 'Suga' in bts:
    bts.remove('Suga')
print(bts)  # ["V", "J-Hope"]

bts = ["V", "J-Hope", "Suga", "Jungkook"]
last_member = bts.pop()  # Son elemanı sil ve döndür 'Jungkook'
print(last_member)  # Jungkook
print(bts)  # ["V", "J-Hope", "Suga"]

# del ile silme
bts = ["V", "J-Hope", "Suga", "Jungkook"]
del bts[0]  # İlk elemanı sil
print(bts)  # ["J-Hope", "Suga", "Jungkook"]

# SyntaxError örneği - del yanlış kullanımı
try:
    bts[0].del  # Hatalı kullanım
except SyntaxError as e:
    print("Hata:", e)
```

## 3. Liste Dilimleme (Slicing)
Dilimleme, listenin belirli bir bölümünü almak için kullanılır. `liste[başlangıç:bitiş:adım]` formatında kullanılır. Bitiş indeksi dahil değildir.


```python
a_list = [10, 20, 30, 40, 50, 60, 70, 80]

# Başlangıç ve bitiş belirtme
print(a_list[1:5])   # [20, 30, 40, 50] - indeks 1'den 4'e kadar
print(a_list[0:5])   # [10, 20, 30, 40, 50] - indeks 0'dan 4'e kadar

# Bitiş belirtmeden
print(a_list[1:])    # [20, 30, 40, 50, 60, 70, 80] - indeks 1'den sona kadar

# Başlangıç belirtmeden
print(a_list[:5])    # [10, 20, 30, 40, 50] - baştan indeks 4'e kadar

# Tüm listeyi kopyalama
print(a_list[:])     # [10, 20, 30, 40, 50, 60, 70, 80] - tüm liste

# Negatif indeks ile dilimleme
print(a_list[-7:-2]) # [20, 30, 40, 50, 60] - sondan 7. elemandan sondan 2. elemana kadar
print(a_list[-7:])  # [20, 30, 40, 50, 60, 70, 80] - sondan 7. elemandan sona kadar
print(a_list[:-2])  # [10, 20, 30, 40, 50, 60] - baştan sondan 2. elemana kadar

# Adım (step) ile dilimleme
letters = ['A', 'B', 'C', 'D', 'E', 'F']
print(letters[2:5])    # ['C', 'D', 'E'] - indeks 2'den 4'e kadar
print(letters[:3])     # ['A', 'B', 'C'] - baştan indeks 2'ye kadar
print(letters[3:])      # ['D', 'E', 'F'] - indeks 3'ten sona kadar
print(letters[:])       # ['A', 'B', 'C', 'D', 'E', 'F'] - tüm liste
print(letters[::2])     # ['A', 'C', 'E'] - 2'şer adımla
print(letters[::-1])    # ['F', 'E', 'D', 'C', 'B', 'A'] - ters sırada
```

## 4. in ve not in Operatörleri
`in` operatörü bir elemanın listede olup olmadığını kontrol eder. `not in` ise olmadığını kontrol eder.


```python
a_list = [10, 20, 30, 40]

# in operatörü
print(10 in a_list)      # True
print(50 in a_list)      # False

# not in operatörü
print(10 not in a_list)  # False
print(50 not in a_list)  # True

# for döngüsü ile kullanım
for n in a_list:
    print(n, end=" ")
print()

# Güvenli silme örneği
n_list = [11, 22, 33, 44, 55, 66]
print(80 in n_list)  # False - 80 listede yok
print(55 in n_list)  # True - 55 listede var

if 55 in n_list:
    n_list.remove(55)
if 88 in n_list:
    n_list.remove(88)
print(n_list)  # [11, 22, 33, 44, 66]
```

## 5. Tuple (Demet)
Tuple, listeye benzer ancak değiştirilemez (immutable) bir veri tipidir. Parantez `()` ile tanımlanır. Değerleri değiştirilemez, bu yüzden sabit veriler için kullanılır.


```python
# Tuple tanımlama
colors = ("red", "green", "blue")
print(colors)  # ('red', 'green', 'blue')

numbers = (1, 2, 3, 4, 5)
print(numbers)  # (1, 2, 3, 4, 5)

# Tuple değiştirilemez - TypeError
try:
    t1 = (1, 2, 3, 4, 5)
    t1[0] = 100  # Hata: tuple değiştirilemez
except TypeError as e:
    print("Hata:", e)
```

## 6. Liste Metodları ve Fonksiyonları
Listeler için kullanılan yerleşik fonksiyonlar ve metodlar: `len()`, `max()`, `min()`, `sum()`, `sort()`, `reverse()`, `index()`, `count()` vb.


#### Liste Metodları Tablosu

| Metod | Fonksiyon |
|-------|-----------|
| `index(x)` | x elemanının indeksini bulur |
| `append(x)` | Listenin sonuna x elemanını ekler |
| `count(x)` | Listedeki x elemanının sayısını döndürür |
| `extend([x1, x2])` | Listenin sonuna [x1, x2] listesini ekler |
| `insert(index, x)` | Belirtilen indeks konumuna x elemanını ekler |
| `remove(x)` | Listedeki x elemanını siler |
| `pop(index)` | Belirtilen indeksteki elemanı siler ve döndürür. İndeks belirtilmezse son elemanı siler ve döndürür |
| `sort()` | Listeyi artan sırada sıralar. `reverse=True` parametresi ile azalan sırada sıralar |
| `reverse()` | Listenin elemanlarını ters sıraya çevirir |


```python
n_list = [200, 700, 500, 300, 400]

# Yerleşik fonksiyonlar
print(len(n_list))  # 5 - uzunluk
print(max(n_list))  # 700 - maksimum değer
print(min(n_list))  # 200 - minimum değer
print(sum(n_list))  # 2100 - toplam

# any() - herhangi biri True mu?
print(any(n_list))  # True - tüm elemanlar True (0 değil)
a_list = [0, '']
print(any(a_list))  # False - tüm elemanlar False

# sort() - sıralama
# Varsayılan olarak artan sıralama yapar
# Liste değişkeninden sonra nokta (.) ve "sort" yazılır
list1 = [20, 10, 40, 50, 30]
list1.sort()  # Artan sıralama (increasing value)
print(list1)  # [10, 20, 30, 40, 50]

list1.sort(reverse=True)  # Azalan sıralama (decreasing value)
print(list1)  # [50, 40, 30, 20, 10]

# reverse() - ters çevirme
letters = ['A', 'B', 'C', 'D', 'E', 'F']
letters.reverse()
print(letters)  # ['F', 'E', 'D', 'C', 'B', 'A']

# index() - elemanın indeksini bulma
n_list = [11, 22, 33, 44, 55, 66]
print(n_list.index(33))  # 2

# count() - eleman sayısını bulma
n_list = [11, 22, 33, 22, 44, 22]
print(n_list.count(22))  # 3
```

Liste işlemleri ile en kısa/en uzun string bulma, iç içe döngüler ile çarpım işlemleri gibi uygulamalar.


```python
# Q1: İlk asal sayıyı yazdırma
prime_list = [2, 3, 5, 7, 11, 13, 17, 19, 23]
print('1st element of prime_list:', prime_list[0])

# Q2: append() ile eleman ekleme
prime_list = [2, 3, 5, 7]
print('Prime numbers :', prime_list)
prime_list.append(11)
print('Prime numbers after addition :', prime_list)

# Q3: İç içe döngü ile çarpım
list1 = [3, 5, 7]
list2 = [2, 3, 4, 5, 6]
for i in list1:
    for j in list2:
        print('{} * {} = {}'.format(i, j, i * j))

# En kısa string bulma
s_list = ['abc', 'bcd', 'bcdefg', 'abba', 'cddc', 'opq']
shortest = s_list[0]
for i in s_list:
    if len(i) < len(shortest):
        shortest = i
print('The shortest string :', shortest)

# En uzun string bulma
s_list = ['abc', 'bcd', 'bcdefg', 'abba', 'cddc', 'opq']
longest = s_list[0]
for i in s_list:
    if len(i) > len(longest):
        longest = i
print('The longest string :', longest)

# En kısa stringleri bulma (birden fazla olabilir)
s_list = ['abc', 'bcd', 'bcdefg', 'abba', 'cddc', 'opq']
s_list.sort(key=len)  # Uzunluğa göre sırala
shortest_len = len(s_list[0])
shortest_list = [i for i in s_list if len(i) == shortest_len]
print('The shortest string :', shortest_list)

# Let's Code - Liste önemi
# Liste olmadan (verimsiz)
height1 = 178.9
height2 = 173.5
height3 = 166.1
height4 = 164.3
height5 = 176.4

# Liste ile (verimli)
heights = [178.9, 173.5, 166.1, 164.3, 176.4]
print(heights)

# Liste oluşturma yöntemleri
bts = ['V', 'Jungkook', 'Jimin']
print(bts)

# append() ile
bts = []
bts.append('V')
print(bts)

# + operatörü ile
bts = []
bts = bts + ["V"]
print(bts)

# Liste elemanlarına erişim
letters = ['A', 'B', 'C', 'D', 'E', 'F']
print(letters[0])  # 'A'
print(letters[1])  # 'B'
print(letters[2])  # 'C'
print(letters[-1])  # 'F' - son eleman

# En kısa string bulma (farklı yöntem)
s_list = ['abc', 'bcd', 'bcdefg', 'abba', 'cddc', 'opq']
for i in s_list:
    if len(i) < len(s_list[0]):
        s_list[0] = i
print('The shortest string :', s_list[0])

# En uzun string bulma (farklı yöntem)
s_list = ['abc', 'bcd', 'bcdefg', 'abba', 'cddc', 'opq']
for i in s_list:
    if len(i) > len(s_list[0]):
        s_list[0] = i
print('The longest string :', s_list[0])

# En kısa stringleri bulma (sort ile)
s_list = ['abc', 'bcd', 'bcdefg', 'abba', 'cddc', 'opq']
s_list.sort(key=len)
for i in s_list:
    if len(i) == len(s_list[0]):
        print(i)

# Liste comprehension ile
s_list = ['abc', 'bcd', 'bcdefg', 'abba', 'cddc', 'opq']
s_list.sort(key=len)
print('The shortest string :', [i for i in s_list if len(i) == len(s_list[0])])
```
