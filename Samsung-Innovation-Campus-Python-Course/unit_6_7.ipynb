## Unit 6 - Conditional Statement-1 : Conditions and Decision Making

Bu ünite, koşullu ifadeler (`if`, `elif`, `else`), girintileme (indentation) kuralları ve karar verme yapılarını kapsar.


#### Mission - Piyango Makinesi
Kullanıcıdan 3 haneli bir sayı alınır ve her hanenin kazanan numaralarla eşleşip eşleşmediği kontrol edilir. Üç hane de eşleşirse kullanıcı kazanır.


```python
# Piyango makinesi örneği
n1, n2, n3 = 2, 3, 9

n = 0
a, b, c = input("Enter a three-digit lottery number : ").split()
a = int(a)
b = int(b)
c = int(c)

if a == n1 or a == n2 or a == n3:  # Eğer bir sayı kazanan numaralardan biriyle eşleşirse
    n += 1                          # n'yi artır
if b == n1 or b == n2 or b == n3:
    n += 1
if c == n1 or c == n2 or c == n3:
    n += 1

if n == 3:  # Eğer n 3 ise, üç sayı da eşleşti
    print("You won the lottery")
```

#### 1. if Koşullu İfadeleri
`if` ifadesi bir koşul kontrol eder; koşul True ise altındaki blok çalıştırılır. Python'da bloklar girintileme (indentation) ile belirlenir. Girintileme hatası `IndentationError` oluşturur.


```python
# Temel if örneği
age = 18
if age < 20:
    print('youth discount')

# Girintileme önemlidir
age = 18
if age < 20:
    print('youth discount')
print('Welcome')  # if bloğunun dışında, her zaman çalışır

# Koşul False olduğunda
age = 24
if age < 20:
    print('youth discount')  # Bu satır çalışmaz
print('Welcome')  # Bu satır çalışır

# Girintileme hatası örneği (yorum satırı olarak)
# age = 18
# if age < 20:
# print('youth discount')  # IndentationError: expected an indented block
```

#### 2. if-elif-else Yapısı
`elif` birden fazla koşulu sırayla kontrol etmek için, `else` ise hiçbir koşul sağlanmadığında çalışacak kodu belirtmek için kullanılır. İkinci dereceden denklem çözümü gibi problemlerde kullanılır.


```python
# İkinci dereceden denklem çözümü
import math

a = int(input('Enter a: '))
b = int(input('Enter b: '))
c = int(input('Enter c: '))

delta = (b * b) - 4 * a * c
if delta < 0:
    print('No root')
elif delta > 0:
    print('There are 2 roots')
    root1 = (- b + (math.sqrt(delta))) / (2 * a)
    root2 = (- b - (math.sqrt(delta))) / (2 * a)
    print('Root1: ', root1, 'Root2: ', root2)
elif delta == 0:
    print('Root1 = Root2')
    root1 = root2 = - b / (2 * a)
    print('Root1 = Root2: ', root1)
```

#### 3. split() Metodu ve Çoklu Giriş
`split()` metodu string'i belirtilen ayırıcıya göre böler ve liste döndürür. Varsayılan ayırıcı boşluktur. Birden fazla değeri tek satırda almak için kullanılır.


```python
# split() metodu örnekleri
str1 = input('Enter three integers divided by space: ')
print('Input: ', str1)
print('Input separated by split() method (default - space): ', str1.split())

# Virgül ile ayırma
str1 = input('Enter three integers divided by comma: ')
print('Input: ', str1)
print('Input separated by split(","): ', str1.split(','))
```

## Unit 7 - Conditional Statement-2

Bu ünite, `else` ve `elif` kullanımı, iç içe koşullu ifadeler (nested if), random modülü ve karmaşık koşullu ifadeleri kapsar.


#### Mission - Penaltı Atışı Oyunu
Bilgisayar rastgele bir savunma pozisyonu seçer, kullanıcı saldırı pozisyonu seçer. Eğer eşleşirse saldırı başarısız, değilse başarılı olur.


```python
# Penaltı atışı oyunu
import random

n = random.randint(1, 5)
if n == 1:
    computer_choice = "Left"
elif n == 2:
    computer_choice = "Middle"
elif n == 3:
    computer_choice = "Right"
elif n == 4:
    computer_choice = "Top Left"
elif n == 5:
    computer_choice = "Top Right"

user_choice = input("Which side will you attack?(left, middle, right, top left, top right) : ")
if computer_choice == user_choice:
    print("Offense failed.")
else:
    print("Congrats!! Offense succeeded.")
print('computer defense position :', computer_choice)
```

#### 1. else ve elif Kullanımı
`else` ifadesi koşul False olduğunda çalışır. `elif` birden fazla koşulu sırayla kontrol eder. `if-else` yapısı iki yönlü karar verme sağlar.


```python
# else kullanımı
score = int(input("Enter score: "))
if score > 90:
    print('Accepted')
    print('Also receive a scholarship')
else:
    print('Not accepted.')

# if-else ile saat kontrolü
hour = 10
if hour < 12:
    print('It is morning')
else:
    print('It is afternoon')

# Çift/tek sayı kontrolü
num = 10
if num % 2 == 0:
    print(num, 'is an even number.')
else:
    print(num, 'is an odd number.')
```

#### 2. İç İçe Koşullu İfadeler (Nested if-else)
Bir `if-else` bloğu başka bir `if-else` bloğunun içine yerleştirilebilir. Bu yapı karmaşık koşulları kontrol etmek için kullanılır.


```python
# İç içe if-else - pozitif/negatif/sıfır kontrolü
num = int(input("Enter an integer: "))
if num >= 0:
    if num == 0:
        print("It is 0.")
    else:
        print("It is positive.")
else:
    print("It is negative.")

# İç içe if-else - çift/tek ve negatif kontrolü
num = 100
if num < 0:
    print(num, 'is negative.')
else:
    print(num, 'is not negative.')
    if num % 2 == 0:
        print(num, 'is an even number.')
    else:
        print(num, 'is an odd number.')

# elif ile aynı örnek (daha temiz)
num = int(input("Enter an integer: "))
if num > 0:
    print("It is positive.")
elif num == 0:
    print("It is 0.")
else:
    print("It is negative.")
```

#### 3. Not Sistemi Örnekleri
Aynı problemi çözmek için farklı yaklaşımlar: birden fazla `if`, iç içe `if-else` veya `if-elif-else`. `if-elif-else` yapısı en temiz ve verimli yaklaşımdır.


```python
# Not sistemi - Birden fazla if kullanımı
score = int(input("Enter score: "))
if score >= 90:
    grade = 'A'
if score < 90 and score >= 80:
    grade = 'B'
if score < 80 and score >= 70:
    grade = 'C'
if score < 70 and score >= 60:
    grade = 'D'
if score < 60:
    grade = 'F'
print('Your grade is :', grade)

# Not sistemi - İç içe if-else kullanımı
score = int(input("Enter score: "))
if score >= 90:
    grade = 'A'
else:
    if score >= 80:
        grade = 'B'
    else:
        if score >= 70:
            grade = 'C'
        else:
            if score >= 60:
                grade = 'D'
            else:     
                if score < 60:
                    grade = 'F'
print('Your grade is :', grade)

# Not sistemi - if-elif-else kullanımı (en temiz)
score = int(input("Enter score: "))
if score >= 90:
    grade = 'A'
elif score >= 80:
    grade = 'B'
elif score >= 70:
    grade = 'C'
elif score >= 60:
    grade = 'D'
else:
    grade = 'F'
print('Your grade is :', grade)
```

#### 5. random Modülü
`random` modülü rastgele sayılar üretmek için kullanılır. `random.randint(a, b)` a ile b arasında (dahil) rastgele tam sayı, `random.randrange(n)` 0 ile n-1 arasında rastgele sayı üretir.


```python
# random modülü örnekleri
import random

# 1 ile 100 arasında rastgele sayı
number = random.randint(1, 100)
print("Random number (1-100):", number)

# 0 ile 1 arasında rastgele sayı (coin toss için)
number = random.randrange(2)
print("Random number (0 or 1):", number)

# Yazı tura oyunu
print("Start coin tossing game.")
coin = random.randrange(2)
if coin == 0:
    print("Front.")
else:
    print("Back.")
print("Game finished.")
```

#### 6. Hesap Makinesi Uygulaması
Kullanıcıdan işlem seçimi ve sayılar alınarak temel matematiksel işlemler yapılır. `eval()` fonksiyonu da kullanılabilir ancak güvenlik açısından dikkatli kullanılmalıdır.


```python
# Hesap makinesi uygulaması
print("1) Addition   2) Subtraction   3) Multiplication   4) Division")
choice = int(input("Enter the desired number of operation : "))
if choice <= 4:
    num1, num2 = input("Enter two numbers for operation.").split()
    num1 = int(num1)
    num2 = int(num2)
    if choice == 1:
        print(num1, "+", num2, "=", num1 + num2)
    elif choice == 2:
        print(num1, "-", num2, "=", num1 - num2)
    elif choice == 3:
        print(num1, "*", num2, "=", num1 * num2)
    elif choice == 4:
        print(num1, "/", num2, "=", num1 / num2)
else:
    print("Entered an incorrect number.")

# Not: eval() kullanımı (güvenlik açısından dikkatli kullanılmalı)
# print(eval(input()))  # Tek satırda hesap makinesi
```
