## Mission - Cafe Menu Program

Kafe menüsü programı: menüye ürün ekleme, fiyat sorgulama ve çıkış işlemleri.


```python
print('Start café menu program... Press q to exit')
cafe_menu = {
    'Ice coffee' : 3000
}

while True:
    str = input('$ ')
    if str.startswith('q'):
        break
    command = str[0]

    if command == '<':
        str = str.replace('<', "")
        inputStr = str.split(':')
        if len(inputStr) < 2:
            print('input error')
            continue
        else:
            cafe_menu[inputStr[0].strip()] = inputStr[1].strip()

    elif command == '>':
        str = str.replace('>', "")
        inputStr = str.strip()
        if inputStr in cafe_menu:
            print(cafe_menu[inputStr])
        else:
            print('{} not in the menu.'.format(inputStr))
    elif command == 'q':
        break
    else:
        print('input error')
print('exiting cafe menu program.')
```

## 1. Add and Edit Key-Value Together

Dictionary'de anahtar-değer çiftlerini ekleme ve düzenleme işlemleri.


#### 1.1. Add, edit dictionary data type

Dictionary'de aynı anahtar kullanılarak hem ekleme hem de düzenleme yapılabilir.


#### 1.2. Dictionary code without setdefault

`setdefault()` kullanmadan dictionary'de eleman sayma örneği.


```python
str1 = ['a', 'b', 'c', 'b', 'a', 'b', 'c']
dic = {}
for ch in str1:
    if ch not in dic.keys():  # Anahtar yoksa ekle
        dic[ch] = 0
    dic[ch] += 1  # Sayacı artır

print('alphabet counting :', dic)  # {'a': 2, 'b': 3, 'c': 2}
```

#### 1.3. Dictionary code with setdefault

`setdefault()` metodu kullanarak daha kısa ve okunabilir kod yazma.


```python
str1 = ['a', 'b', 'c', 'b', 'a', 'b', 'c']
dic = {}
for ch in str1:
    dic.setdefault(ch, 0)  # Anahtar yoksa 0 değeriyle ekle, varsa değiştirme
    dic[ch] += 1  # Sayacı artır
    print(list(dic.items()))  # Her adımda dictionary durumunu göster

print('alphabet counting :', dic)  # {'a': 2, 'b': 3, 'c': 2}
```

```python
# setdefault() örneği: Anahtar yoksa ekle, varsa değiştirme
dic = {
    'a' : 10,
    'b' : 20,
    'c' : 30,
    'd' : 40
}
dic.setdefault('f', 50)  # 'f' anahtarı yok, 50 değeriyle ekle
print(dic)  # {'a': 10, 'b': 20, 'c': 30, 'd': 40, 'f': 50}
```

#### 1.4. Update changes value of key

`update()` metodu dictionary'deki mevcut anahtarın değerini değiştirir.


```python
dic = {
    'a' : 10,
    'b' : 20,
    'c' : 30,
    'd' : 40
}
dic.update(a = 50)  # 'a' anahtarının değerini 50 olarak güncelle
print(dic)  # {'a': 50, 'b': 20, 'c': 30, 'd': 40}
```

#### 1.5. When update method adds key that does not exist

`update()` metodu mevcut olmayan bir anahtarı ekler.


```python
dic.update(e = 50)  # 'e' anahtarı yok, ekle
print(dic)  # {'a': 50, 'b': 20, 'c': 30, 'd': 40, 'e': 50}
```

#### 1.6. Modify multiple items with update method

`update()` metodu ile birden fazla anahtar-değer çiftini tek seferde güncelleme.


```python
dic.update(a = 900, f = 60)  # Birden fazla anahtarı güncelle
print(dic)  # {'a': 900, 'b': 20, 'c': 30, 'd': 40, 'e': 50, 'f': 60}
```

#### 1.7. Precautions for using the update method

`update()` metodu string olmayan anahtarlar için dictionary kullanılmalıdır.


```python
dic2 = {
    1: 'one',
    3: 'three'
}
# String olmayan anahtarlar için dictionary kullan
dic2.update({1: 'ONE', 3: 'THREE'})
print(dic2)  # {1: 'ONE', 3: 'THREE'}
```

#### 1.8. How to modify the update method key when it's not a string

String olmayan anahtarlar için `update()` metodunu liste veya tuple ile kullanma.


```python
# Liste ile update (iç içe liste: [[anahtar, değer], ...])
dic2.update([[2, 'TWO'], [4, 'FOUR']])
print(dic2)  # {1: 'ONE', 3: 'THREE', 2: 'TWO', 4: 'FOUR'}
```

```python
# Tuple ile update (iç içe tuple: ((anahtar, değer), ...))
dic2.update(((2, 'TWO'), (4, 'FOUR')))
print(dic2)  # {1: 'ONE', 3: 'THREE', 2: 'TWO', 4: 'FOUR'}
```
