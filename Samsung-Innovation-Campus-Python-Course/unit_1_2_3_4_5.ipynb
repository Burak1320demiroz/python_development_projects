Bu hücre, `print` fonksiyonunun metin biçimlendirme, sayısal işlemler, `sep` ve `end` parametreleri ile hata yönetimine etkisini tek akış içinde görselleştirir. Hata örneklerinde `TypeError` mesajlarını yakalayarak programın durmasını engelliyoruz; böylece aynı çıktı bloğunda tüm örnekler izlenebiliyor.


```python
# Çok satırlı metin
print('''Hello

World''')
print()

# Sayısal işlemler
print(10 + 7)
print(10 - 7.5)
print(7.5)
print()

# Satır sonu karakteri kaçırma
print('''Hello\
World''')
print()

# Birden fazla argümanla print
print("Hello", "World!")
print(10, 20)
print("Hello", 10)
print()

# Birleştirme ve toplama
print("Hello" + "World!")
print(10 + 20)
try:
    print("Hello" + 10)
except TypeError as exc:
    print("TypeError:", exc)
print()

# Çarpma işlemleri
print("Hello" * 3)
print(10 * 20)
try:
    print("Hello" * "World!")
except TypeError as exc:
    print("TypeError:", exc)
print()

# end ve sep parametreleri
print("Hello", "test", end="")
print("World!")
print("Hello", end=",")
print("World!")
print("Hello", "World!", "Test", sep="")
print("Hello", "World!", 2, sep="#")
print()

# Kişisel tanıtım
print("Hello! I will introduce myself.")
print("Name: Penh Polydet")
print("Age: 20")
print("Job: Student")
print("Address: Phnom Penh, Cambodia")
print("Place of Birth: Phnom Penh, Cambodia")
print()

# Ayırıcı çizgiler
print("*************************************")
print("-------------------------------------")
print()

# Runtime error örneği (simülasyon)
print("Runtime Error")
try:
    a, b = "ten", "2"
    result = int(a) / int(b)
    print('{}/{} = {}'.format(a, b, result))
except Exception:
    print('Check if the integers are correctly entered.')
print()

# Elmas deseni çizdirme
print("Print Diamond")
print("     ", "*")
print("    ", "*" * 3)
print("   ", "*" * 5)
print(" ", "*" * 9)
print("   ", "*" * 5)
print("    ", "*" * 3)
print("     ", "*")
```

## Unit 3 - Basic of Numeric Data Type and Arithmetic Operation

Bu ünite, sayısal veri tipleri, aritmetik işlemler, operatör önceliği ve veri tipi dönüşümlerini kapsar. Daire alan ve çevre hesaplama örneği ile pratik uygulama yapılır.


```python
# Daire alan ve çevre hesaplama
PI = 3.14
radius = float(input("Enter radius : "))
area = PI * radius * radius
circum = 2.0 * PI * radius

print('Radius of the circle : ', radius)
print('Area of the circle : ', area)
print('Circumference of the circle : ', circum)
```

#### 1. Aritmetik Operatörler
Python'da temel matematiksel işlemler için kullanılan operatörler: toplama (+), çıkarma (-), çarpma (*), bölme (/), tam bölme (//), kalan (%) ve üs alma (**).


```python
# Aritmetik operatör örnekleri
print("Toplama (+):", 100 + 20)
print("Çıkarma (-):", 100 - 20)
print("Çarpma (*):", 100 * 20)
print("Bölme (/):", 100 / 20)
print("Tam bölme (//):", 100 // 20)
print("Kalan (%):", 100 % 20)
print("Üs alma (**):", 10 ** 2)
```

#### 2. Veri Tipleri ve type() Fonksiyonu
Python'da temel veri tipleri: int (tam sayı), float (ondalıklı sayı), str (metin), bool (True/False), list, tuple, dict, complex (karmaşık sayı), None. `type()` fonksiyonu bir değişkenin veri tipini döndürür.


```python
# Veri tipleri örnekleri
print("int:", type(100))
print("float:", type(100.0))
print("str:", type('100'))
print("complex:", type(10 + 3j))
print("bool:", type(True))
print("list:", type([10, 20, 30]))
print("tuple:", type((10, 20, 30)))
print("dict:", type({'name': 'David', 'age': 23}))
print("None:", type(None))

# Hata örneği - string ile sayısal işlem
try:
    result = '10' / 2
except TypeError as e:
    print("Hata:", e)
```

#### 3. Operatör Önceliği
Python'da operatörler belirli bir öncelik sırasına göre işlenir. Parantez içindeki işlemler önce yapılır, sonra üs alma, çarpma/bölme, en son toplama/çıkarma gelir.


```python
# Operatör önceliği örnekleri
print("10 + 20 * 30 =", 10 + 20 * 30)  # Çarpma önce yapılır
print("(10 + 20) * 30 =", (10 + 20) * 30)  # Parantez önce yapılır
```

#### 4. Veri Tipi Dönüşümü
`int()`, `float()`, `str()` gibi fonksiyonlar veri tiplerini birbirine dönüştürmek için kullanılır. String ile sayısal işlem yapmak için önce dönüşüm yapılmalıdır.


```python
# Veri tipi dönüşümü örnekleri
print("int('10') / 2 =", int('10') / 2)  # String'i int'e çevirip bölme
print("'I like number ' + str(10) =", 'I like number ' + str(10))  # Sayıyı string'e çevirip birleştirme
```

#### 5. String Operatörleri
String'lerde `+` operatörü birleştirme yapar, `*` operatörü tekrar ettirir. `*` operatörü string ile kullanılırken sağ tarafta mutlaka tam sayı olmalıdır, aksi halde TypeError oluşur.


```python
# String operatörleri örnekleri
print("'I' + ' Love' + ' Python' =", 'I' + ' Love' + ' Python')
print("'Python ' * 10 =", 'Python ' * 10)
print("'*' * 50 =", '*' * 50)
print("str(100) * 10 =", str(100) * 10)

# Hata örneği - float ile çarpma
try:
    result = '*' * 3.1
except TypeError as e:
    print("Hata (string * float):", e)
```

## Unit 4 - Variables and Inputs

Bu ünite, değişken atama yöntemleri, birleşik atama operatörleri, kullanıcıdan veri alma ve pratik uygulamaları kapsar.


#### 1. Değişken Atama Yöntemleri
Python'da aynı anda birden fazla değişkene değer atanabilir (simultaneous assignment) veya aynı değer birden fazla değişkene atanabilir (multiple assignment). Değişken sayısı ile değer sayısı eşit olmalıdır.


```python
# Simultaneous assignment (eşzamanlı atama)
x, y = 100, 200
result = x + y
print("x, y = 100, 200 sonucu:", result)

# Multiple assignment (çoklu atama)
num1 = num2 = num3 = 200
print("num1 = num2 = num3 = 200:", num1, num2, num3)

# Hata örnekleri
try:
    x, y, z = 100, 200  # Değer sayısı yetersiz
except ValueError as e:
    print("Hata (değer sayısı yetersiz):", e)

try:
    x, y = 100, 200, 300  # Değer sayısı fazla
except ValueError as e:
    print("Hata (değer sayısı fazla):", e)

try:
    300 = 300  # Literal'a atama yapılamaz
except SyntaxError as e:
    print("Hata (literal'a atama): SyntaxError")
```

#### 2. Birleşik Atama Operatörleri
Birleşik atama operatörleri (`+=`, `-=`, `*=`, `/=`) değişkenin değerini güncellemek için kısa yoldur. `num += 10` ifadesi `num = num + 10` ile aynıdır.


```python
# Birleşik atama operatörleri
num = 200
num += 100
print("num += 100 sonucu:", num)
num -= 100
print("num -= 100 sonucu:", num)
num *= 20
print("num *= 20 sonucu:", num)
num /= 2
print("num /= 2 sonucu:", num)
```

#### 3. Input Fonksiyonu ve Kullanımı
`input()` fonksiyonu kullanıcıdan string olarak veri alır. Sayısal işlemler için `int()` veya `float()` ile dönüşüm yapılmalıdır. `isdigit()` metodu string'in sayı olup olmadığını kontrol eder.


```python
# String input örneği
name = input("Enter Name: ")
print("Hello,", name + ".")
print("Welcome to the World of Python Programming.")

# Integer input örneği
x = int(input("Enter first integer: "))
y = int(input("Enter second integer: "))
s = x + y
print("The sum of", x, "and", y, "is", s)

# isdigit() ile kontrol
x = input("Enter the first integer: ")
print('Input Data Type:', type(x))
if x.isdigit() == True:
    x = int(x)
    print('Input Data Type:', type(x))
else:
    print('x is not a number type.')
```

#### 4. Pratik Uygulamalar
BMI hesaplama, para üstü hesaplama, dikdörtgen alanı, Pisagor teoremi gibi gerçek hayat problemlerini çözmek için değişkenler ve aritmetik operatörler kullanılır.


```python
# BMI hesaplama
weight = float(input("Enter weight in kg: "))
height = float(input("Enter height in m: "))
bmi = (weight / (height ** 2))
print("Your BMI=", bmi)

# Para üstü hesaplama
money = int(input("Money Paid : "))
price = int(input("Price of Item : "))
change = money - price
print("Amount of Change : ", change)
coin500s = change // 500
change = change % 500
coin100s = change // 100
print("Pieces of 500 won coins: ", coin500s)
print("Pieces of 100 won coins: ", coin100s)

# Dikdörtgen alanı
width = 30
height = 60
areaOfRectangle = width * height
print("Area of Rectangle :", areaOfRectangle)

# Pisagor teoremi
a = int(input("Enter a : "))
b = int(input("Enter b : "))
c = (a**2 + b**2) ** 0.5
print("Hipotenüs:", c)

# Daire alan ve çevre (geliştirilmiş)
PI = 3.141592
radius = int(input("Enter radius of a circle: "))
area = PI * radius * radius
circum = 2.0 * PI * radius
print('Radius of the circle : ', radius)
print('Circumference of the circle :', circum, 'Area of the circle :', area)

# Kare tablosu
n = 2
print("a\t\tn\t\ta**n")
for a in range(2, 7):
    result = a**n
    print(a, "\t       ", n, "\t       ", result)
```

#### 5. Değişkenler ve Veri Tipleri
Değişkenler veri saklamak için kullanılır ve her değişkenin bir veri tipi vardır. Yanlış veri tipi kullanımı TypeError'a neden olur. String ile sayısal işlem yapmadan önce dönüşüm yapılmalıdır.


```python
# Değişken kullanımı
radius = 4.0
print('Radius of Circle', radius)
print('Area of Circle', 3.14 * radius * radius)
print('Circumference of Circle', 2.0 * 3.14 * radius)

# Değişken ve veri tipi
name = 'David'
age = 27
print(type(name))
print(type(age))

# TypeError örneği
try:
    my_age = 22
    my_height = '177'
    my_age += 1
    my_height += 1  # String'e int eklenemez
    print(my_age, my_height)
except TypeError as e:
    print("Hata (TypeError):", e)

# Sıfıra bölme hatası
try:
    num1 = 3
    num2 = 0
    print(num1 / num2)
except ZeroDivisionError as e:
    print("Hata (ZeroDivisionError): Sıfıra bölme yapılamaz")
```

## Unit 5 - Logic and Comparison Operations

Bu ünite, karşılaştırma operatörleri, mantıksal operatörler ve boolean veri tipini kapsar. Artık yıl hesaplama gibi pratik uygulamalar yapılır.


#### 1. Karşılaştırma Operatörleri
Karşılaştırma operatörleri (`>`, `<`, `==`, `!=`, `>=`, `<=`) iki değeri karşılaştırır ve boolean (True/False) değer döndürür. String'lerde de kullanılabilir ve alfabetik sıraya göre karşılaştırma yapar.


```python
# Karşılaştırma operatörleri
print('Result of 10 > 5:', 10 > 5)
print('Result of 5 < 1:', 5 < 1)
print('Result of 5 == 5:', 5 == 5)
print('Result of 5 != 5:', 5 != 5)
print("Result of 'a' > 'b':", 'a' > 'b')

a = 100
b = 200
print("a == b:", a == b)
print("a != b:", a != b)
print("a > b:", a > b)
print("a < b:", a < b)
print("a >= b:", a >= b)
print("a <= b:", a <= b)

# String karşılaştırma
print("'aaa' == 'aaa':", 'aaa' == 'aaa')
print("'aaa' == 'bbb':", 'aaa' == 'bbb')
print("'aaa' != 'aaa':", 'aaa' != 'aaa')
print("'aaa' != 'bbb':", 'aaa' != 'bbb')

# Çift sayı kontrolü
n = int(input('Enter the integer: '))
print('Is the integer even?', n % 2 == 0)
```

#### 2. Boolean Veri Tipi ve bool() Fonksiyonu
Boolean veri tipi sadece `True` ve `False` değerlerini alır. `bool()` fonksiyonu herhangi bir değeri boolean'a dönüştürür. 0, None, boş string ve boş liste False, diğerleri True döndürür.


```python
# Boolean veri tipi
print("bool(1):", bool(1))
print("bool(0):", bool(0))
print("bool(True):", bool(True))
print("bool(False):", bool(False))

# Farklı değerlerin boolean karşılığı
print("bool(-1):", bool(-1))  # 0 dışındaki sayılar True
print("bool(0):", bool(0))  # 0 False
print("bool(None):", bool(None))  # None False
print("bool(''):", bool(''))  # Boş string False
print("bool('hello'):", bool('hello'))  # Dolu string True
print("bool([]):", bool([]))  # Boş liste False
print("bool([10, 20]):", bool([10, 20]))  # Dolu liste True
```

#### 3. == ve is Operatörleri Arasındaki Fark
`==` operatörü değerleri karşılaştırır, `is` operatörü ise aynı nesneye referans edip etmediğini kontrol eder. `in` operatörü bir değerin başka bir değer içinde olup olmadığını kontrol eder.


```python
# == vs is farkı
a = 1
b = 1.0
print("a == b (değer karşılaştırma):", a == b)  # True - değerler eşit
print("a is b (nesne karşılaştırma):", a is b)  # False - farklı veri tipleri

# in operatörü
print("'aaa' in 'aaa-bbb-ccc':", 'aaa' in 'aaa-bbb-ccc')
print("'bbb' in 'aaa-bbb-ccc':", 'bbb' in 'aaa-bbb-ccc')
print("'ddd' in 'aaa-bbb-ccc':", 'ddd' in 'aaa-bbb-ccc')

# Zincirleme karşılaştırma
n = 100
print('n =', n)
print('n is between 0 and 200:', 0 < n < 200)

# Hata örneği - yanlış operatör sırası
try:
    num1 = 100
    num2 = 200
    num1 =< num2  # Yanlış: <= olmalı
except SyntaxError:
    print("Hata: Operatör sırası yanlış, <= kullanılmalı")
```

#### 4. Mantıksal Operatörler
Mantıksal operatörler (`and`, `or`, `not`) boolean değerlerle çalışır ve karmaşık koşullar oluşturmak için kullanılır. `and` her iki koşul da True ise True döndürür, `or` en az biri True ise True döndürür, `not` boolean değeri tersine çevirir.


```python
# Mantıksal operatörler
x = True
y = False
print("x and y:", x and y)
print("x or y:", x or y)
print("not x:", not x)
print("not y:", not y)

# Pratik örnekler
num = int(input('Enter an Integer: '))
result = (num >= 0 and num <= 100 and num % 2 == 0)
print('Is the input an even integer between 0 and 100?', result)

# Tek sayı kontrolü
n = int(input('Enter an Integer : '))
print('Is the integer odd?::', n % 2 != 0)

# Artık yıl hesaplama
year = int(input("Enter the year: "))
leap_year = (year % 4 == 0 and year % 100 != 0) or year % 400 == 0
print("Leap year:", leap_year)

# 3 basamaklı sayı kontrolü (yüzler basamağı 3 mü?)
n = int(input('Enter a 3-digit integer: '))
print("Hundreds digit is 3:", n // 100 == 3)

# 5'in katı kontrolü
n = int(input('Enter an integer: '))
print("Is multiple of 5:", n % 5 == 0)
```
